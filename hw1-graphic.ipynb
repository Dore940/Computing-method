```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define the function f(x,y) = x^2 - y^2
def f(x, y):
    return x**2 - y**2
```

```python
# Create a 3D grid of points in the domain [0,1]*[0,1]
x = np.linspace(0, 1, 50)
y = np.linspace(0, 1, 50)
X, Y = np.meshgrid(x, y)
```

```python
# Evaluate the function f(x,y) at each point in the grid
Z = f(X, Y)
```

```python
# Create a 3D plot of the function using Matplotlib's mplot3d toolkit
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
plt.show()
```

```python
# Evaluate the function f(x,y) at the edges of the domain [0,1]*[0,1]
x = np.linspace(0, 1, 50)
y = np.zeros(50)
f1 = f(x, y)
```

```python
x = np.zeros(50)
y = np.linspace(0, 1, 50)
f2 = f(x, y)
```

```python
x = np.linspace(0, 1, 50)
y = np.ones(50)
f3 = f(x, y)
```

```python
x = np.ones(50)
y = np.linspace(0, 1, 50)
f4 = f(x, y)
```

```python
# Create 2D plots of the function f(x,y) at the edges of the domain
fig, axs = plt.subplots(2, 2, figsize=(8, 8))
axs[0, 0].plot(x, f1)
axs[0, 0].set_title('f(0,y)')
axs[0, 1].plot(y, f2)
axs[0, 1].set_title('f(x,0)')
axs[1, 0].plot(x, f3)
axs[1, 0].set_title('f(1,y)')
axs[1, 1].plot(y, f4)
axs[1, 1].set_title('f(x,1)')
plt.show()
```
